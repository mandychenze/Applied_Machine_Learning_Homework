```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/Users/zechen/Desktop/AML/Homework/homework-1-mandychenze/task2/2.1data.csv')

data.head()
```

```python
#prepare data
Year = data['Year'].tolist()
swimming = data['Swimming pool drownings'].tolist()
cage = data['Nicholas Cage'].tolist()

#draw line plot
plt.figure(figsize=(8,6))
ax1 = plt.gca()
#plt.tick_params(axis='x', which='top', bottom=True, top=True,color='r',labelbottom=True, labeltop=True)
line1, = ax1.plot(Year, swimming, c='r')
ax2 = ax1.twinx()

line2, = ax2.plot(Year, cage, c='black')
ax1.set_ylabel("Swimming pool drownings")
ax2.set_ylabel("Nicholas Cage")
plt.title('Number of People who drowned by falling into a pool \ncorrelates with \nFilms Nicholas Cage appeared in \nCorrelation: 66.6%(r=0.666004)', y=1.08)
ax2.legend((line1, line2),
           ("Swimming pool drownings", "Nicholas Cage"))

#in order to get an axis on top, draw a new line(line3)
ax3 = ax1.twiny()
line3, = ax3.plot(Year, swimming, c='red')

plt.tight_layout()
plt.savefig('task21.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
mpg = pd.read_csv('/Users/zechen/Desktop/AML/Homework/homework-1-mandychenze/task2/mpg.csv', index_col=0)
mpg.head()
```

```python
displ = mpg['displ'].tolist()
cty = mpg['cty'].tolist()
drv = mpg['drv'].tolist()

displ_fwd = mpg[mpg.drv =='f']['displ'].tolist()
cty_fwd = mpg[mpg.drv =='f']['cty'].tolist()
#dry_fwd = mpg[mpg.drv =='f']['drv'].tolist()

displ_rwd = mpg[mpg.drv =='r']['displ'].tolist()
cty_rwd = mpg[mpg.drv =='r']['cty'].tolist()
#dry_rwd = mpg[mpg.drv =='r']['drv'].tolist()

displ_4wd = mpg[mpg.drv =='4']['displ'].tolist()
cty_4wd = mpg[mpg.drv =='4']['cty'].tolist()
#dry_4wd = mpg[mpg.drv =='4']['drv'].tolist()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
#plt.xticks([], [])
#plt.yticks([], [])
plt.axis('off')

ax = fig.add_subplot(221)
ax.scatter(displ_fwd, cty_fwd, c= "orange", s=70,label="FWD")
ax.scatter(displ_rwd, cty_rwd, c= "blue", s=70, label="RWD")
ax.scatter(displ_4wd, cty_4wd, c= "black", s=70, label="4WD")
ax.legend(loc=0, ncol=1, title = 'drive train')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('displacement (I)')
ax.set_ylabel('fuel economy (mpg)')

ax2 = fig.add_subplot(222)
ax2.scatter(displ_fwd, cty_fwd, c= "orange", alpha=0.3, s=70, label="FWD")
ax2.scatter(displ_rwd, cty_rwd, c= "blue", alpha=0.3, s=70, label="RWD")
ax2.scatter(displ_4wd, cty_4wd, c= "black", alpha=0.3, s=70, label="4WD")
ax2.legend(loc=0,ncol=1, title = 'drive train')
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)
ax2.set_xlabel('displacement (I)')
ax2.set_ylabel('fuel economy (mpg)')

def jitter(num):
    #stdev = (max(num)-min(num))/100  #stdev = arr.max()/100
    stdev = max(num)/100.0 #don't divide by integer, or will be rounded
    return num + np.random.randn(len(num)) * stdev

ax3 = fig.add_subplot(223)
ax3.scatter(jitter(displ_fwd), jitter(cty_fwd), c= "orange", alpha=0.3, s=70,label="FWD")
ax3.scatter(jitter(displ_rwd), jitter(cty_rwd), c= "blue", alpha=0.3, s=70, label="RWD")
ax3.scatter(jitter(displ_4wd), jitter(cty_4wd), c= "black", alpha=0.3, s=70,label="4WD")
ax3.legend(loc=0,ncol=1, title = 'drive train')
ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)
ax3.set_xlabel('displacement (I)')
ax3.set_ylabel('fuel economy (mpg)')

#jitter with greater variance
def jitter2(num):
    stdev = max(num)/20.0
    return num + np.random.randn(len(num)) * stdev

ax4 = fig.add_subplot(224)
ax4.scatter(jitter2(displ_fwd), jitter2(cty_fwd), c= "orange", alpha=0.3, s=70,label="FWD")
ax4.scatter(jitter2(displ_rwd), jitter2(cty_rwd), c= "blue", alpha=0.3, s=70, label="RWD")
ax4.scatter(jitter2(displ_4wd), jitter2(cty_4wd), c= "black", alpha=0.3, s=70,label="4WD")
ax4.legend(loc=0,ncol=1, title = 'drive train')
ax4.spines['right'].set_visible(False)
ax4.spines['top'].set_visible(False)
ax4.set_xlabel('displacement (I)')
ax4.set_ylabel('fuel economy (mpg)')

fig.savefig('task23.png')
plt.show()
```
